```python
from googletrans import Translator
translator = Translator(raise_exception=True)
```

```python
import pandas as pd
```

```python
text = '''definitely stay again and have been recommending it to friends.'''
result = translator.translate(text, dest='bn')
print(result.text)
```

```python
#read the csv 
df = pd.read_csv('train.csv')

df.head()
```

```python
eng_list = df['Description']
eng_list = list(eng_list[100:200])
print(len(eng_list))
```

```python
translations = translator.translate(eng_list, dest='bn')
```

```python
#make a new list
bangla_list = []
for t in translations:
    print(t.text)
    bangla_list.append(t.text)
```

```python
# dictionary of lists  
dict = {'description': bangla_list}  
       
df = pd.DataFrame(dict) 
    
# saving the dataframe 
df.to_csv('100.csv',index=False) 
```
